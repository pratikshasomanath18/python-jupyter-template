### Exercise 4: Pie Chart with Percentages

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['blue', 'red', 'green', 'purple']
explode = (0.1, 0, 0, 0)  # Emphasize Python section

# Error handling
if not isinstance(labels, list) or not isinstance(sizes, list):
    raise TypeError("Both labels and sizes should be lists.")
if len(labels) != len(sizes):
    raise ValueError("Labels and sizes must have the same length.")
if sum(sizes) <= 0:
    raise ValueError("Sum of sizes must be greater than zero.")

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, explode=explode, 
        startangle=140, shadow=True)

# Title
plt.title('Programming Language Popularity')

# Show the plot
plt.show()
```
