### Exercise 4: Pie Chart with Percentages

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['gold', 'lightblue', 'lightcoral', 'lightgreen']
explode = (0.1, 0, 0, 0)  # Highlighting the first slice

# Error Handling
if len(labels) != len(sizes):
    raise ValueError("Labels and sizes lists must be of the same length.")
if len(colors) != len(labels):
    raise ValueError("Colors list must have the same length as labels and sizes.")
if len(explode) != len(labels):
    raise ValueError("Explode list must have the same length as labels and sizes.")

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, explode=explode, startangle=140)

# Title
plt.title('Customized Pie Chart')

# Show the plot
plt.show()
```
