### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np
from collections.abc import Iterable

# Data
x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 1, 4, 9, 16])
categories = ['A', 'B', 'C', 'D']
values = [5, 3, 9, 6]
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [10, 15, 7, 5]

# Error Handling
if not isinstance(x, Iterable) or not isinstance(y, Iterable):
    raise TypeError("Both x and y must be iterable.")
if len(x) != len(y):
    raise ValueError("x and y must have the same length.")
if len(categories) != len(values):
    raise ValueError("Categories and values lists must have the same length.")
if len(labels) != len(sizes):
    raise ValueError("Labels and sizes lists must have the same length.")

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# First subplot: Line plot
axs[0, 0].plot(x, y, marker='o', linestyle='--', color='b', label='Quadratic Growth')
axs[0, 0].set_title('Line Plot')
axs[0, 0].legend()

# Second subplot: Bar plot
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title('Bar Plot')

# Third subplot: Scatter plot
axs[1, 0].scatter(x, y, color='blue', marker='o')
axs[1, 0].set_title('Scatter Plot')

# Fourth subplot: Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightcoral', 'lightgreen'])
axs[1, 1].set_title('Pie Chart')

# Show the plots
plt.show()
```
