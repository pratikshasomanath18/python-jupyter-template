### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
categories = ['A', 'B', 'C', 'D']
values = [5, 3, 9, 6]
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [10, 15, 7, 5]

# Error Handling
if not all(isinstance(i, (int, float)) for i in x + y + values + sizes):
    raise TypeError("All numerical data must be integers or floats.")

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# First subplot: Line plot
axs[0, 0].plot(x, y, marker='o', linestyle='--', color='b')
axs[0, 0].set_title('Line Plot')

# Second subplot: Bar plot
axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Plot')

# Third subplot: Scatter plot
axs[1, 0].scatter(x, y, color='purple')
axs[1, 0].set_title('Scatter Plot')

# Fourth subplot: Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, shadow=True)
axs[1, 1].set_title('Pie Chart')

plt.tight_layout()  # Adjust spacing
plt.show()
```
