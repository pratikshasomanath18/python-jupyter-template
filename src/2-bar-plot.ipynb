### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'orange']  # Each bar has a different color

# Error Handling
if len(categories) != len(values):
    raise ValueError("The 'categories' and 'values' lists must have the same length.")
if len(colors) != len(categories):
    raise ValueError("The 'colors' list must have the same length as 'categories' and 'values'.")

# Create a bar plot
plt.bar(categories, values, color=colors)

# Labels and Title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Customized Bar Plot')

# Show the plot
plt.show()
```
