### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']  # Different colors for each bar

# Error handling
if not isinstance(categories, list) or not isinstance(values, list):
    raise TypeError("Both categories and values should be lists.")
if len(categories) != len(values):
    raise ValueError("Categories and values must have the same length.")

# Create a bar plot
plt.bar(categories, values, color=colors, edgecolor='black')

# Customize the plot
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot of Categories vs. Values')

# Show the plot
plt.show()
```
