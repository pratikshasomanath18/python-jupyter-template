### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt
import numpy as np
from collections.abc import Iterable

# Data
x = [1, 2, 3, 4, 5]
y = [5, 7, 8, 5, 6]
annotations = ['A', 'B', 'C', 'D', 'E']  # Labels for each point

# Error Handling
if not isinstance(x, Iterable) or not isinstance(y, Iterable):
    raise TypeError("Both x and y must be iterable (list, tuple, or NumPy array).")
if len(x) != len(y):
    raise ValueError("x and y must have the same length.")
if len(annotations) != len(x):
    raise ValueError("Annotations list must have the same length as x and y.")

# Create a scatter plot
plt.scatter(x, y, color='blue', marker='o')

# Add annotations
for i in range(len(x)):
    plt.annotate(annotations[i], (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='right')

# Labels and Title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Scatter Plot')

# Show the plot
plt.show()
```
