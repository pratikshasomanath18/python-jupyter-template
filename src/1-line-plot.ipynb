### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt

# Data
x = list(range(0, 11))  # x values from 0 to 10
y = [i**2 for i in x]   # Square of x

# Error handling
if not isinstance(x, list) or not isinstance(y, list):
    raise TypeError("Both x and y should be lists.")
if len(x) != len(y):
    raise ValueError("x and y lists must have the same length.")

# Create a line plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x²')

# Customize plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plot of y = x²')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
