### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np
from collections.abc import Iterable

# Sample Data
x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 1, 4, 9, 16])

# Error Handling
if not isinstance(x, Iterable) or not isinstance(y, Iterable):
    raise TypeError("Both x and y must be iterable (list, tuple, or NumPy array).")

if len(x) != len(y):
    raise ValueError("x and y must have the same length.")

# Plotting
plt.plot(x, y, marker='o', linestyle='--', color='b', label='Quadratic Growth')

# Labels and Title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Customization')

# Legend
plt.legend()

# Show the plot
plt.show()
```
